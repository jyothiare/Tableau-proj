# Which Airport is better to fly out or in ?  EWR,PHL or JFK 


The U.S. Department of Transportation's (DOT) Bureau of Transportation Statistics (BTS) tracks the on-time performance of domestic flights operated by large air carriers. Summary information on the number of on-time, delayed, canceled and diverted flights are available for the public. I have downloaded the raw data for the year 2016 for all the airlines for the tristate area airports EWR,PHL and JFK. I live in the tristate area. The purpose of my analysis is to find out the best airlines and airport to board or disembark a plane to avoid any delays and disrubtions.

The questions I have set to answer are

__1. Which airport have the least arrival and departure delays.__ 
  
__2. Which Airlines have better performance and from which airport?__
 
__3. What are the type of delays high at each airport?__

__4. Which destination has the most departure delays by each airport?__

__5. Which airport sees the most arrival delays and from where?__
    
__6. Which day of the week is best to flyout to avoid delays?__

__7. Which day of the week is best to avoid arrival delays.__
       


  





# Design

Dataset information

This dataset is taken from Rita website. I have downloaded the csv files for each month of 2016 and combined them and then filtered data for the three airports I am intrested in Newark(EWR),Philadelphia(PHL) and Newyork(JFK). I cleaned the CSV files data by removing all the diverted flight information. I have also downloaded the longitude and lattitude for all the airports as csv files and joined with the ONTIME Performance table .

Version 1

As I said earlier I have started this project with a set of questions. I live in the Tristate area and thought this should be an useful visualization for any one in Tristate area looking to minmize the probablity of a delayed flight.
I started by trying to find out how many flights are on time or not for each airport? If the flight was not on time what was the status ? Was it delayed ,Cancelled or diverted? From this I found when a flight was not on time most of the time it is because of a delayed flight. The next step was to drill down to see how each airline carrier measured in terms of flight delays for both Arrivals and Departures. The next step was to find out how Arrival and Departure delays are effected by a destination ? I also wanted to see if I am flying out or flying in which day of the week would minimize my probability of a delay. To represent all of these I considered different types of graphs but with the multiple categories and sub categories I had I found bar charts worked the best.  But I also wanted to show a geographical representation of each airport's domestic arrivals and dipartures by carriers and volume.

# Feedback

Version 2 (based on feedback #1 and own improvements)

 First and foremost I had to make a lot of spelling and sizing adjustments on different screens based on the feedback. Most of the time the tooltip was displaying the column names so had to change them to readable names. Changed labels to make better sense(like changed week day to day of week). The colors I have chosen initially werenot uniform and since I was asked to see if I can make it more consistent I changed the color scheeme. I do think the changed color scheem is better. The one adjustment I didn't make is to change the colors of 7 weekdays. I received the complaint that it is too many colors. But I am not sure what is the alternative other than moving to pastel colors. I felt this as personal choice so left the colors in tact. One other question I have received is instead of having seperate graphs by numbers and percentages why not combine both for arrivals or departures.  The reason I had to keep them seperate is that I can only use one of the measure either the no of flights or the percentage of delayed flights when I am comparing between 3 airports . I can show the other measure in tool tip  but it wouldn't be the same . For example The arrival delays from LAX to JFK is higher by volume when compared to MCO to JFK but when you check the percentage of delays MCO delays are higher. So because of these reasons I have chosen to show these differences by different graphs.
Several of the filters are consolidated and reduced the clutter as much as possible.
